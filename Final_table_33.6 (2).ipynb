```python
import pandas as pd


#reading csv and converting it into pandas dataframe
df = pd.read_csv(r"C:\Users\putta\Downloads\table-33.6.csv")
df
```

```python
 #Specify columns and rows to drop using indexing
df.drop([0,2], inplace = True)

#Drop rows with any NaN values

df.dropna(axis=0, how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)
```

```python
#to make header row using indexing
df.columns = df.iloc[0]
df

```

```python
# Drop the row with index 4
df.drop([4], inplace = True)
df

```

```python
 #Specify columns and rows to drop using indexing
df.drop([5,6,38,47,48,49],inplace = True)


df.columns.values[2]='National Park Area'
df.columns.values[4]='Wildlife Sanctuaries Area'

df
```

```python
# Select the first three columns of the DataFrame 'df' and create a new DataFrame 'data'
data = df[df.columns[:3]]

# Select the columns excluding the first two columns of 'df' and create a new DataFrame 'data2'
data2 = df[[df.columns[0]] + df.columns[3:].to_list()]

# Print the 'data' DataFrame
print(data, data2)
```

```python

# For data DataFrame
data.insert(1, 'category', 'National Parks')
display(data)

# For data2 DataFrame
data2.insert(1, 'category', 'Wildlife Sanctuaries')
display(data2)

# Print the column names
print(data.columns)
print(data2.columns)
```

```python
# Set column names 'number' and 'area' for the last two columns in the 'data' DataFrame
data.columns.values[2:]=['number','area']

# Set column names 'number' and 'area' for the last two columns in the 'data2' DataFrame
data2.columns.values[2:]=['number','area']

# Print the column names of the 'data' DataFrame
print(data.columns)
print(data2.columns)
```

```python
#concating both dataframes 
result = pd.concat([data,data2])
result
```

```python


# Insert 'unit' column with a constant value
result.insert(4, 'unit', 'number in Absolute, area in Sq Km', True)

# Insert 'note' column with an empty string
result.insert(5, 'note', '', True)

# Rename the first column to 'state'
result.rename(columns={result.columns[0]: 'state'}, inplace=True)

# Display the updated DataFrame
result
```

```python
#replacing using dictionaries 
replace_dict = {
    '&': 'and',
    'Total': 'All India',
    'D.and N.Haveli': 'Dadra and Nagar Haveli',
    'A. and N. Islands': 'Andaman and Nicobar Islands',
    '+': '',
    'Chhatisgarh': 'Chhattisgarh'
}

# Strip leading and trailing whitespaces
result['state'] = result['state'].str.strip()

# Display the updated DataFrame
display(result)

```
